### Импортируем необходимые библиотеки

```python
import pandas as pd
```

##### Загрузка CSV файла в DataFrame

```python
# Импорт CSV файла
# ---------------------------------------------------------------
# Загрузка CSV файла в DataFrame.


df_csv = pd.read_csv('countries_dataset/countries-of-the-world.csv', encoding = 'utf-8', sep = ',')

# Используем файл 'countries-of-the-world.csv' который лежит в каталоге с ноутбуком,
# encoding='utf-8': Этот аргумент указывает кодировку файла. UTF-8 поддерживает множество языков, включая кириллицу.
# если с кириллицей проблемы - применяем кодировку 'cp1251', 
# разделитель полей в файле ';' (точка с запятой)
#  read_csv по умолчанию считывает первую строку как заголовки



# Выводим первые 5 строк DataFrame
df_csv.head()

# Выводим последние 5 строк DataFrame
# df_csv.tail()
```

##### Загрузка JSON файла в DataFrame

```python
# ---------------------------------------------------------------
# Импорт JSON файла
# ---------------------------------------------------------------
# Загрузка JSON файла в DataFrame
df_json = pd.read_json('countries_dataset/countries-table.json')

# Сохраняем в Excel
df_json.to_excel('countries_dataset/countries-table.xlsx')

# Выводим первые 5 строк DataFrame
df_json.head()

# Выводим последние 5 строк DataFrame
# df_json.tail()
```

##### Загрузка Excel файла в DataFrame

```python
# ---------------------------------------------------------------
# Импорт Excel файла
# ---------------------------------------------------------------
# Загрузка Excel файла в DataFrame. Мы используем первый лист в файле Excel.
df_excel = pd.read_excel('countries_dataset/countries-table.xlsx', sheet_name=0)

# Выводим первые 5 строк DataFrame
df_excel.head()

# Выводим последние 5 строк DataFrame
# df_excel.tail()
```
